```python
# Import Required Liberies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataset
df = pd.read_csv("student_marks.csv")  
df.head()
```

```python
# Understand The Data
df.shape
df.columns
df.info()
```

```python
# Data Cleaning
df.isnull().sum()
```

```python
# Check Duplicates
df.duplicated().sum()
```

```python
# Encode Gender
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
```

```python
# Statistical Summary
df.describe()
```

```python
# Univariate Analysis
plt.hist(df['Math'], bins=5)
plt.xlabel("Math Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Math Scores")
plt.show()
```

```python
sns.boxplot(y=df['Attendance'])
plt.title("Attendance Distribution")
plt.show()
```

```python
# Bivariate Analysis
plt.scatter(df['Math'], df['Science'])
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Math vs Science Performance")
plt.show()
```

```python
plt.scatter(df['Attendance'], df['English'])
plt.xlabel("Attendance")
plt.ylabel("English Marks")
plt.title("Attendance vs English")
plt.show()
```

```python
# Correlation Analysis
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Gender-Wise Performance
df.groupby('Gender')[['Math','Science','English']].mean()
```
